```python
%matplotlib inline
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler #(converting datas into z-score)


from sklearn.decomposition import PCA #(all this belong to a family of mathematical transformation called Matrix Decompositions)
from scipy.stats import zscore #(same like StandardScaler)

#Read the data set

mpg_df=pd.read_csv("D:\Machine Learning\Machine Learning Projects\PCA(Car mpg -Supervised)\car-mpg.csv")

# Drop the car name column(non-numerical) as it is useless for the model
car_name=mpg_df['car_name']
mpg_df=mpg_df.drop('car_name', axis=1)
mpg_df.head()
```

```python
# horsepower is an object type though it is supposed to be numeric(has 6 "?" in the hp column)

#To check all the rows in the column are digit
temp=pd.DataFrame(mpg_df.hp.str.isdigit()) # if the string is made of digits store True else False in the hp column
temp[temp['hp']==False] # from temp take only those rows where hp has false
```

```python
#CORRECTING THE ROWS WITH "?" IN THE HP COLUMN

#On inspecting records number 32,126,...374, we find "?" inn the columns 
#Replace them with "nan" and remove the records from the data frame that have "nan"
mpg_df= mpg_df.replace('?',np.nan)
mpg_df= mpg_df.apply(lambda x: x.fillna(x.median()),axis=0)


#converting the hp column from object/string type to float
mpg_df['hp']=mpg_df['hp'].astype('float64')
```

```python
#Visually inspect the covariance between independent dimensions and mpg
sns.pairplot(mpg_df,diag_kind='kde')
```

```python
#FOR COVARIANCE MATRIX(FOR THE ENTIRE DATA INCLUDING TARGET VARIABLE)
from scipy.stats import zscore
mpg_z=mpg_df.apply(zscore)
#creating covariance matrix
cov_matrix= np.cov(mpg_z.T)#(np.cov-->This function will multiply mpg_z and its transpose,ie. mpg_z * mpg_Z.T)
print('Covariance Matrix \n%s',cov_matrix)
```

```python
#COVARIANCE MATRIX USING STANDARDSCALER
#sc = StandardScaler()
#mpg_df_std =  sc.fit_transform(mpg_df)          
#cov_matrix = np.cov(mpg_df_std.T)
#print('Covariance Matrix \n%s', cov_matrix)
```

```python
#splitting out the dependent variable and the independent variable

X=mpg_df[mpg_df.columns[1:-1]]
y=mpg_df["mpg"]
```

```python
# COVARIANCE MATRIX ONLY FOR THE INDEPENDENT VARIABLE

# We transform (centralize) the entire X(independent variable data) to zscores  through transformation.  
#We will create the PCA dimensions on this distribution. 
#sc=StandardScaler()
#X_std=sc.fit_transform(X)
#cov_matrix= np.cov(X_std.T)

X_z = X.apply(zscore)
cov_matrix= np.cov(X_z.T)

print('Covariance Matrix \n%s',cov_matrix)
```

```python
# Pair plot for the independent variable(X)
sns.pairplot(X,diag_kind='kde')
```

```python
# APPLYING EIGEN TRANSFORMATION (LINEAR ALGEBRA FUNCTION ) TO THE COVARIANCE MATRIX
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix) #here only conceptually the mathematical space is getting rotated.
print('Eigen Vectors \n%s', eigenvectors) #no.of eigen vectors=  no.of independent variable
print('\n Eigen Values \n%s', eigenvalues)
```

```python
#MAKE A SET OF EIGEN PAIRS(EIGEN VALUE,EIGEN VECTORS) AND SORT IT(DESCENDING ORDER)
eig_pairs=[(eigenvalues[index],eigenvectors[:,index]) for index in range(len(eigenvalues))]

# sorting the EIGEN PAIRS(ASCENDING ORDER)
eig_pairs.sort() # .sort() choose the first column(eigen values) and sort it(ascending order)

#Changing Ascending order to DESCENDING ORDER 
eig_pairs.reverse()
print(eig_pairs)


#Extract the descending ordered eigenvalues and eigenvectors
eigvalues_sorted= [eig_pairs[index][0] for index in range(len(eigenvalues))]
eigvectors_sorted= [eig_pairs[index][1] for index in range(len(eigenvalues))]

#lets confirm our sorting worked, print out eigenvalues
print('Eigenvalues in descending order: \n%s' %eigvalues_sorted)
```

```python
tot = sum(eigenvalues)
var_explained = [(i / tot) for i in sorted(eigenvalues, reverse=True)]  # an array of variance explained by each eigen vector 
# there will be 8 entries as there are 8 eigen vectors)
cum_var_exp = np.cumsum(var_explained)  # an array of cumulative variance. There will be 8 entries with 8 th entry...
#... cumulative reaching almost 100%
```

```python
#PLOT A GRAPH Between Eigen vectors and Cumulative Eigen values
plt.bar(range(1,8),var_explained, alpha=0.5, align='center', label='individual explained variance')
plt.step(range(1,8),cum_var_exp, where='mid', label='cumulative explained variance')
plt.ylabel('Explained Variance Ratio')
plt.xlabel('Principal Components')
plt.legend(loc='best')
plt.show()
```

```python
#REDUCING MATHEMATICAL SPACE
# p_reduce represents reduced mathematical space....

p_reduce = np.array(eigvectors_sorted[0:4]) # Reducing from 8 to 4 dimension space
X_z_4D= np.dot(X_z,p_reduce.T)#Mapping the X_z(Z-scored X) into the new 4 Dimensions(projecting ...
                              #.....original data into principal component dimensions)
proj_data_df = pd.DataFrame(X_z_4D)# converting array to dataframe for pairplot

```

```python
from sklearn import model_selection

test_size = 0.30 # taking 70:30 training and test set
seed = 7 # Random number seeding for repeatability of the code
X_train,X_test,y_train,y_test = model_selection.train_test_split(proj_data_df, y, test_size=test_size, random_state=seed)
```

```python
# PAIR PLOT BETWEEN THESE 4 DIMENSIONS(4 PCS) IN THE SQUARE MATRIX
sns.pairplot(proj_data_df,diag_kind='kde')
```

```python
#Let us build a linear regression model on the PCA dimensions

#Import Linear Regression Machine Learning Library
from sklearn.linear_model import LinearRegression

regression_model=LinearRegression()
regression_model.fit(X_train,y_train)

regression_model.coef_
```

```python
#model intercept
regression_model.intercept_
```

```python
regression_model.score(X_test,y_test)
```
